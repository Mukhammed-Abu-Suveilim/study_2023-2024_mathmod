```julia
using Plots
using DifferentialEquations

x0 = 22022
y0 = 33033

a = 0.343 
b = 0.895 
c = 0.699 
h = 0.433

P(t) = 2 * sin(2*t)
Q(t) = 2 * cos(t)

manpower = [x0, y0]
p = (a, b, c, h)
T = [0, 1]

function F(du, u, p, t)
    a, b, c, h = p
    du[1] = -a * u[1] - b * u[2] + P(t) 
    du[2] = -c * u[1]*u[2] - b * u[2] + Q(t)
end

prob = ODEProblem(F, manpower, T, p)
```

```julia
sol = solve(prob)
```

```julia
plot(sol)
```
